```python
def findFirstPeak(wf):
    """
    Finds where/around where the first peak occurs. Since we're integrating the power, it doesn't matter if it's
    not the real peak position.
    
    Returns:
    peak[0]: int
    bin where the first peak happens
    """
    normWF = wf/max(abs(wf))
    analytic_signal = scipy.signal.hilbert(normWF)
    amplitude_envelope = np.abs(analytic_signal)
    peaks,_ = scipy.signal.find_peaks(amplitude_envelope, width=2, prominence=0.5)
    return peaks[0]
```

```python
findFirstPeak(wf.v)
```
