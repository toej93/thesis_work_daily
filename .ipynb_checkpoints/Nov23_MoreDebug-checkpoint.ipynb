```python
import pyrex.custom.ara as ara
theta = np.radians(67.83)
phi = np.radians(119.79)
antH = ara.HpolAntenna(name="Dummy Hpol", position=(0, 0, 0), power_threshold=0)
antV = ara.VpolAntenna(name="Dummy Vpol", position=(0, 0, 0), power_threshold=0)
polarization=np.array([-np.sin(phi),np.cos(phi),-1/np.sin(theta)])
# polarization=np.array([0,0,1])
fft_f = np.linspace(0,1000,1001)
```

```python
dir_resH = antH.antenna.directional_response(theta=theta, phi=phi, polarization=polarization)(fft_f*1E6)
dir_resV = antV.antenna.directional_response(theta=theta, phi=phi, polarization=polarization)(fft_f*1E6)
```

```python
plt.figure(figsize=(8,5))
plt.plot(fft_f,abs(dir_resV), label="Vpol")
plt.plot(fft_f,abs(dir_resH), label="Hpol")
plt.ylabel("Gain")
plt.xlabel("Freq [MHz]")
plt.grid()
plt.legend()
plt.title("Antenna gain at $\phi=%0.2f$, $\\theta=%0.2f$"%(np.degrees(phi),np.degrees(theta)))
```

```python
heffH = antH.antenna.frequency_response(fft_f*1E6)
response_antennaH = dir_resH*heffH

heffV = antV.antenna.frequency_response(fft_f*1E6)
response_antennaV = dir_resV*heffV
```

```python
plt.figure(figsize=(8,5))
plt.plot(fft_f,abs(response_antennaV), label="Vpol")
plt.plot(fft_f,abs(response_antennaH), label="Hpol")
plt.ylabel("$abs(H_{eff})$ [m]")
plt.xlabel("Freq [MHz]")
plt.grid()
plt.legend()
plt.title("Antenna eff. height at $\phi=%0.2f$, $\\theta=%0.2f$"%(np.degrees(phi),np.degrees(theta)))
```
