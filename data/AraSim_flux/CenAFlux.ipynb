```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
flux = pd.read_csv("Flux_CenA.txt", skiprows=1)#From arxiv 2102.00213
Energy = np.log10(flux.Energy)
E2flux = np.log10(flux.flux*1E-10/(3600*24*365)*1E-9)
```

```python
# E2Flux = flux.flux*1E-10/(3600*24*365)*1E-9
plt.plot(flux.Energy, E2flux)
```
